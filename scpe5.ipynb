```python
%%writefile modelo.mod
param n; #cantidad de centros
param m; #cantidad de regiones

# Sets
set N := 1..n; # Set of centros de servicios
set M := 1..m; # Set of regiones

#parametros
param a{M,N} >=0; # matriz que establece si la region i puede ser servida por el centro j
param c{N} >=0; #costo por instalar centro j

#variables
var x{N} >=0, binary; #variable que toma el valor de 1 si el centro j se instala, cero en caso contrario

# Objective function
minimize costo: sum{i in N} c[i] * x[i];

# Constraints

# Cada región debe ser cubierta
subject to aseguracubrimiento {i in M}:
sum{j in N} a[i,j] * x[j] >= 1;




```

```python

import numpy as np
import requests

# Función para descargar y leer el archivo SCP desde la web
def leer_scp_archivo_desde_web(url):
    respuesta = requests.get(url)
    contenido = respuesta.text.splitlines()

    # Eliminar espacios y saltos de línea
    lineas = [linea.strip() for linea in contenido if linea.strip()]

    # 1. Extraer número de centros (n) y número de regiones (m)
    num_variables, num_restricciones = map(int, lineas[0].split())

    # 2. Extraer costos de los centros (param c)
    costos = []
    indice_linea = 1
    while len(costos) < num_restricciones:
        costos.extend(map(int, lineas[indice_linea].split()))
        indice_linea += 1

    # 3. Extraer la matriz de cobertura (param a)
    matriz_cobertura = np.zeros((num_variables, num_restricciones), dtype=int)

    for i in range(num_variables):
        datos = list(map(int, lineas[indice_linea].split()))
        cantidad_centros = datos[0]  # Cuántos centros cubren esta región
        centros_cubren = datos[1:]   # Índices de los centros que cubren

        # Marcar en la matriz las variables que cubren la restricción
        for centro in centros_cubren:
            matriz_cobertura[i, centro - 1] = 1  # -1 porque Python indexa desde 0

        indice_linea += 1

    return num_restricciones, num_variables, costos, matriz_cobertura

# Ejemplo de uso
url = 'https://people.brunel.ac.uk/~mastjjb/jeb/orlib/files/scpe5.txt'
num_restricciones, num_variables, costos, matriz_cobertura = leer_scp_archivo_desde_web(url)

# Mostrar resultados en el formato del usuario
print(f'param n = {num_restricciones};  # Number of centers')
print(f'param m = {num_variables};  # Number of regions')

print('\nparam c :=')
for i, costo in enumerate(costos, 1):
    print(f'{i} {costo}')
print(';')

print('\nparam a :', ' '.join(str(i) for i in range(1, num_restricciones + 1)), ':=')
for i in range(num_variables):
    fila = ' '.join(str(matriz_cobertura[i, j]) for j in range(num_restricciones))
    print(f'{i+1} {fila}')
print(';')
```
